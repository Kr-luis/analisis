```python
#11.	Exportar a MySQL el archivo unificado.
import pandas as pd
from sqlalchemy import create_engine #necesario para exportar a mysql pip install sqlalchemy
final_csv = 'final.csv'
# Con
usuario = 'root'
contrasena = '123456'
host = 'localhost'
puerto = '3306'
nombre_bd = 'repaso2'

df = pd.read_csv(final_csv)
#conexión a la base de datos mysql con sqlalchemy
engine = create_engine(f'mysql+mysqlconnector://{usuario}:{contrasena}@{host}:{puerto}/{nombre_bd}')

nombre_tabla = 'tabla_final'

# Exportar el df a mysql
df.to_sql(name=nombre_tabla, con=engine, if_exists='replace', index=False)

print(f'Datos exportados a la tabla {nombre_tabla} en MySQL.')

```
